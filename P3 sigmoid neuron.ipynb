## P3  Sigmoid neuron
Als eerst maken we een sigmoid functie. Die zie je boven de class staan. Deze functie wordt in de class gebruikt bij de activatie methode. Ook heeft de class Neuron een methode str die de neuron overzichtelijk print.

```python
from tabulate import tabulate
```

```python
def sigmoid(z):
    e = 2.71828
    F = 1/(1 + e**(-z))
    return F
```

```python
class Neuron(object):

    def __init__(self, bias, weights):
        self.bias = bias
        self.weights = weights
        
        
    def __str__(self):
        # print 
        return 'Neuron(Bias=' + str(self.bias) + ', Weights=' + str(self.weights) + ', inputs=' + str(self.inputs)
    

    def activate(self, inputs):
        total = 0
        self.inputs = inputs
        
        for i in range(len(self.inputs)):
            total += self.inputs[i] * self.weights[i]
        
        total = total + self.bias
        
        return sigmoid(total)
    
    
```

## Neuronlayer & NeuronNetwerk
### Neuronlayer

Net als bij de perceptron maken we nu voor de neuron een layer en een netwerkclass aan. Dat gebeurt hieronder

```python
class Neuronlayer(object):
    
    def __init__(self, neurons):
        self.neurons = neurons     
    
    
    def activate(self, inputs):
        total = []
        
        #ittereert over de neurons
        for i in range(len(self.neurons)):
            # Saved welke neuron er aan de beurt is
            neuron = self.neurons[i]   
            
            #Neemt de input en voert de activate functie uit met de waardes die die neuron heeft
            total.append(neuron.activate(inputs))
              
        
        return total
```

### NeuronNetwerk

```python
class NeuronNetwork(object):
    
    def __init__(self, neuronlayers):
        self.neuronlayers = neuronlayers     
    
    
    def feed_foward(self, inputs):
        # Loopt over de layers heen en houdt bij, bij welke layer hij is.
        for i in range(len(self.neuronlayers)):
            neuronlayer = self.neuronlayers[i]   
            
            # output is de uitkomst van de neuronlayer activatie functie de gegeven inputs
            output = neuronlayer.activate(inputs)
            
            # Maakt van de output de nieuwe input voor de volgende iteratie.
            inputs = output
            
        return output
```

## Half adder
Nu we de layers en het netwerk kunnen maken is het tijd om een Half Adder te maken. Die bestaat uit twee layers. Layer1 is opgebouwd uit een NandGate, een OrGate en een AndGate. Bestaat uit een AndGate en een Gate die de waarde van de vorige AndGate doorgeeft. Eerst maken we een NandGate en die tunen we ook gelijk

```python
inputs= [[0, 0], [0,1], [1,0], [1,1]]
results = []

# Maakt een neuron aan met de juiste waardes om de gate te simuleren en geeft de input daar aan mee
NandGate = Neuron(6 , [-4.5, -4.5])

# Loopt over de inputs
for i in range(len(inputs)):
    pair = []

    
    # Maakt een paar van de input en output
    pair.append(inputs[i])
    pair.append(NandGate.activate(inputs[i]))
    results.append(pair)
                
print(tabulate(results, headers=['Input', 'Output']))
```

Nu we weten dat de NandGate werkt kunnen we de HalfAdder maken. Door de NandGate te gebruiken en neurons die we eerder gemaakt hebben.

```python
# Hiddenlayer1 bestaat uit een NandGate, Orgate en AndGate
Hiddenlayer1 = Neuronlayer([Neuron(6 , [-4.5, -4.5]), Neuron(-2.5 , [4.5, 4.5]), Neuron(-8 , [5, 5])])

# Hiddenlayer2 bestaat uit een gate die de waarde van de andgate door geeft en een Andgate die alleen de waardes
# van de Nand en OrGate gebruikt
Hiddenlayer2 = Neuronlayer([Neuron(-8 , [5, 5, 0]),Neuron(-0.5 , [0, 0, 1])])
```

```python
network = NeuronNetwork([Hiddenlayer1, Hiddenlayer2])
inputs = [[0,0],[0,1],[1,0],[1,1]]
results = []

# Loopt over de inputs en maakt pairs net als in het vorige tests
for i in range(len(inputs)):
    pair = []
    
    # Voegt de input toe aan de lijst pair
    pair.append(inputs[i])
    
    # Voegt de output van het netwerk toe aan de lijst pair
    pair.append(network.feed_foward(inputs[i]))
    
    # Voegt het volledige paar toe aan de lijst results
    results.append(pair)

    
print(tabulate(results, headers=['Input', 'sum/carry']))
```

Je ziet dat de waardes van de sum bij [0,1] en [1,0] het hoogst zijn en de carry van [1,1] is het hoogst. Dat is de gewenste uitkomst want die moeten normaal 1 zijn maar het zit niet heel dicht bij 1. Ik denk dat dat komt door de doorgeef neuron.
