## Perceptron ##
Eerst maken we een class Percptron, die kun je de bias, weigthts en een input meegeven. Hij springt over wanneer de som hoger of gelijk is aan 0.

```python
class Perceptron(object):

    def __init__(self, bias, weights, inputs):
        self.bias = bias
        self.weights = weights
        self.inputs = inputs
        
    def __str__(self):
        # print 
        return 'Perceptron(Bias=' + str(self.bias) + ', Weights=' + str(self.weights) + ', inputs=' + str(self.inputs)
    
    
    def step_function(self):
        # Houdt bij wat de som is om over de threshhold te komen.
        total = 0
        
        #ittereert over de inputs en de weights en verminigvuldigd ze.
        for i in range(len(self.inputs)):
            total += self.inputs[i] * self.weights[i]
        
        # bij het totaal wordt de bias opgeteld en checkt vervolgens of hij wel/niet aan moet gaan. 
        total = total + self.bias
        if total >= 0:
            act = 1
        else:
            act = 0
            
        return act
```

Hier onder zijn een paar tests geinitialiseerd om te testen of de perceptron juist werkt. Eerst een Invertgate, daarna een Andgate en tot slot een Orgate. De output en de bijbehorende input worden samen gereturnd in een lijst.
## A) ##

```python
inputs= [[0], [1]]
results = []

#loopt over de inputs
for i in range(len(inputs)):
    pair = []
    
    #Maakt een perceptron aan met de juiste waardes om de gate te simuleren en geeft de input daar aan mee
    InvertGate = Perceptron(0.5 , [-0.51], inputs[i])
    
    #Maakt een paar van de input en output
    pair.append(inputs[i])
    pair.append(InvertGate.step_function())
    results.append(pair)
results
```
