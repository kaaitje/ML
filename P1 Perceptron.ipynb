## Perceptron ##
Eerst maken we een class Percptron, die kun je de bias, weigthts en een input meegeven. Hij springt over wanneer de som hoger of gelijk is aan 0.

```python
from tabulate import tabulate
```

```python
class Perceptron(object):

    def __init__(self, bias, weights):
        self.bias = bias
        self.weights = weights
        
        
    def __str__(self):
        # print 
        return 'Perceptron(Bias=' + str(self.bias) + ', Weights=' + str(self.weights) + ', inputs=' + str(self.inputs)
    
    
    def activate(self, inputs):
        # Houdt bij wat de som is om over de threshhold te komen.
        total = 0
        self.inputs = inputs
        #ittereert over de inputs en de weights en verminigvuldigd ze.
        for i in range(len(self.inputs)):
            total += self.inputs[i] * self.weights[i]
        
        # bij het totaal wordt de bias opgeteld en checkt vervolgens of hij wel/niet aan moet gaan. 
        total = total + self.bias
        if total >= 0:
            act = 1
        else:
            act = 0
            
        return act
```

## Perceptronlayer & Network##

### Perceptronlayer ###
Hieronder maak ik als eerst een perceptronlayer aan. Die krijgt een lijst als input en output een lijst met evenveel elementen als perceptrons in die laag.

```python
class Perceptronlayer(object):
    
    def __init__(self, perceptrons):
        self.perceptrons = perceptrons     
    
    
    def activate(self, inputs):
        total = []
        
        #ittereert over de perceptrons
        for i in range(len(self.perceptrons)):
            # Saved welke perceptron er aan de beurt is
            perceptron = self.perceptrons[i]   
            
            #Neemt de input en voert de activate functie uit met de waardes die die perceptron heeft
            total.append(perceptron.activate(inputs))
              
        
        return total
```

### PerceptronNetwork
Nu kan ik een perceptron netwerk aanmaken door meerder layers aan elkaar te koppelen. Deze class heeft een functie feed_foward die de output van de vorige layer gebruikt als input voor de volgende itteratie. Tot slot return hij de laatste output

```python
class PerceptronNetwork(object):
    
    def __init__(self, perceptronlayers):
        self.perceptronlayers = perceptronlayers     
    
    
    def feed_foward(self, inputs):
        
        for i in range(len(self.perceptronlayers)):
            perceptronlayer = self.perceptronlayers[i]   
            output = perceptronlayer.activate(inputs)
            inputs = output
            
        return output
```

## XorGate

een Xorgate bestaat uit een laag met een NandGate en een OrGate. De output van die laag gaat door naar de tweede laag die bestaat uit een AndGate. De gewenste output van een XorGate is 1 als er een van de inputs juist is. Als beide onjuist of juist zijn moet de output 0 zijn. Daarvoor moet ik wel eerst een NandGate maken en testen, dat doe ik als eerst.

```python
inputs= [[0, 0], [0,1], [1,0], [1,1]]
results = []

# Maakt een perceptron aan met de juiste waardes om de gate te simuleren en geeft de input daar aan mee
NandGate = Perceptron(1 , [-1, -1])

# Loopt over de inputs
for i in range(len(inputs)):
    pair = []

    
    # Maakt een paar van de input en output
    pair.append(inputs[i])
    pair.append(NandGate.activate(inputs[i]))
    results.append(pair)
                
print(tabulate(results, headers=['Input', 'Output']))
```

Nu ik een NandGate heb kan ik de layers maken die nodig zijn voor het netwerk. De OrGate en AndGate had ik al eerder gemaakt en getest dus die gebruik ik gewoon opnieuw.

```python
# Hiddenlayer1 bestaat uit een NandGate en OrGate
Hiddenlayer1 = Perceptronlayer([Perceptron(1 , [-1, -1]), Perceptron(-0.5 , [0.5, 0.5])])

# Hiddenlayer2 bestaat uit een AndGate
Hiddenlayer2 = Perceptronlayer([Perceptron(-1 , [0.5, 0.5])])
```

Nu ik de layers heb is het simpel om een test te schrijven voor de XorGate, dat doe ik hier onder.

```python
#Initialiseert het netwerk, de outputs en een lijst om de resultaten in op te slaan.
network = PerceptronNetwork([Hiddenlayer1, Hiddenlayer2])
inputs = [[0,0],[0,1],[1,0],[1,1]]
results = []

# Loopt over de inputs en maakt pairs net als in het vorige tests
for i in range(len(inputs)):
    pair = []
    
    # Voegt de input toe aan de lijst pair
    pair.append(inputs[i])
    
    # Voegt de output van het netwerk toe aan de lijst pair
    pair.append(network.feed_foward(inputs[i]))
    
    # Voegt het volledige paar toe aan de lijst results
    results.append(pair)

    
print(tabulate(results, headers=['Input', 'Output']))
```

## Half adder

Een half adder bestaat uit twee layers, de eerste layer bestaat uit een nandgate, een orgate en een andgate. De tweede layer bestaat uit een gate die de waarde van alleen de Andgate doorgeeft. De AndGate verwerkt alleen de waardes van de Nand en OrGate. Wat er dus uit de laatste Andgate komt is de sum. Wat er doorgegeven wordt is de carry.

```python
# Hiddenlayer1 bestaat uit een NandGate, Orgate en AndGate
Hiddenlayer1 = Perceptronlayer([Perceptron(1 , [-1, -1]), Perceptron(-0.5 , [0.5, 0.5]), Perceptron(-1 , [0.5, 0.5])])

# Hiddenlayer2 bestaat uit een gate die de waarde van de andgate door geeft en een Andgate die alleen de waardes
# van de Nand en OrGate gebruikt
Hiddenlayer2 = Perceptronlayer([Perceptron(-1 , [0.5, 0.5, 0]),Perceptron(-0.5 , [0, 0, 1])])
```

```python
network = PerceptronNetwork([Hiddenlayer1, Hiddenlayer2])
inputs = [[0,0],[0,1],[1,0],[1,1]]
results = []

# Loopt over de inputs en maakt pairs net als in het vorige tests
for i in range(len(inputs)):
    pair = []
    
    # Voegt de input toe aan de lijst pair
    pair.append(inputs[i])
    
    # Voegt de output van het netwerk toe aan de lijst pair
    pair.append(network.feed_foward(inputs[i]))
    
    # Voegt het volledige paar toe aan de lijst results
    results.append(pair)

    
print(tabulate(results, headers=['Input', 'sum/carry']))
```
